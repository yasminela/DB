```python
import numpy as np
import pandas as pd

# Load data
data=pd.read_csv('house-prices.csv')

data.head()
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
```

```python
plt.scatter(data["area"], data["price"], color="gold")
plt.xlabel("Area (in Square Feet)")
plt.ylabel("Price (in $)")
```

```python
# spare les variables indpendantes (features) et dpendantes (prix)
#x :datafrime
#y:target
# plus de donnes pour l'apprentissage que le test
X = data.drop("price", axis=1)
y = data["price"]
```

```python
from sklearn.linear_model import LinearRegression
# cre le moddle de rgression linaire
model = LinearRegression()
# ajuste le modle saux donne d'apprentissage
model.fit(X, y)
```

```python
# Prdictions du modle
y_pred = model.predict(X)
```

```python
# Visualisation
plt.scatter(X, y, color="blue", alpha=0.6, label="Donnes relles")
plt.plot(X, y_pred, color="red", linewidth=2, label="Rgression linaire")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Prdire le prix dune maison")
plt.legend()
plt.show()
```

```python
# effectue une prdiction pour un nouvel ensemble de features
new_house= [[1500]]
```

```python
prix= model.predict(new_house)
print("Prix attendu pour la nouvelle maison : ", prix)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error (MSE) :", mse)
print("Mean Absolute Error (MAE) :", mae)
print("Coefficient de dtermination R :", r2)
```
